```python
import pandas as pd
```

```python
stock_df = pd.read_html( 'http://kind.krx.co.kr/corpgeneral/corpList.do?method=download&searchType=13', header=0)[0]
stock_df['종목코드'] = stock_df['종목코드'].map(lambda x: f'{x:0>6}')
corp_list = stock_df[['회사명','종목코드','업종','지역']]
code_list = corp_list['종목코드'].values
```

```python
import dart_fss
from dart_fss import get_corp_list
api_key='960a281f210548c84eb6c951ec3a10587b1ede65' 
dart_fss.set_api_key(api_key=api_key)
corp_list = dart_fss.corp.get_corp_list()
```

```python
code_list1 = code_list[1:]
code_list1
```

```python
# 기업 개별재무제표 
a = corp_list.find_by_stock_code('066570')
fs_separate = a.extract_fs(bgn_de='20190101', separate=True)
```

```python
# fs_box
fs = fs_separate[0].fillna(method='ffill', axis=1)
class3 = fs.iloc[:,6].values
last = fs['20191231'].values
last = last.reshape(-1)
dict={'계정':list(class3),a.corp_name:list(last)}
fs_box = pd.DataFrame(dict)

# cis_box
cis = fs_separate[2].fillna(method='ffill', axis=1)
class3 = cis.iloc[:,6].values
last = cis['20190101-20191231'].values
last = last.reshape(-1)
dict={'계정':list(class3),a.corp_name:list(last)}
cis_box = pd.DataFrame(dict)

# cf_box
cf = fs_separate[3].fillna(method='ffill', axis=1)
class3 = cf.iloc[:,5].values
last = cf['20190101-20191231'].values
last = last.reshape(-1)
dict={'계정':list(class3),a.corp_name:list(last)}
cf_box = pd.DataFrame(dict)
```

```python
for i in code_list :    
    try :
        ## 개별 재무제표
        a = corp_list.find_by_stock_code(i)
        fs_separate = a.extract_fs(bgn_de='20190101', separate=True)

        ## 재무상태표
        fs = fs_separate[0].fillna(method='ffill', axis=1)
        class3 = fs.iloc[:,6].values
        last = fs['20191231'].values
        last = last.reshape(-1)
        dict={'계정':list(class3),a.corp_name:list(last)}
        fs1 = pd.DataFrame(dict)
        fs_box = pd.merge(fs_box,fs1,how='left',on='계정')

        ## 포괄손익계산서
        cis = fs_separate[2].fillna(method='ffill', axis=1)
        class3 = cis.iloc[:,6].values
        last = cis['20190101-20191231'].values
        last = last.reshape(-1)
        dict={'계정':list(class3),a.corp_name:list(last)}
        cis1 = pd.DataFrame(dict)
        cis_box = pd.merge(cis_box,cis1,how='left',on='계정')

        ## 현금흐름표
        cf = fs_separate[3].fillna(method='ffill', axis=1)
        class3 = cf.iloc[:,5].values
        last = cf['20190101-20191231'].values
        last = last.reshape(-1)
        dict={'계정':list(class3),a.corp_name:list(last)}
        cf1 = pd.DataFrame(dict)
        cf_box = pd.merge(cf_box,cf1,how='left',on='계정')
        
    except : continue
```
