```python
# 결측치 처리-보간법 이용
data = data.interpolate()
# 나머지 결측치-0으로 채워넣음
data = data.fillna(0)
```

```python
np.save("data_company.npy",arr = data)
data = np.load("data_company.npy", allow_pickle=True)
```

```python
# integer변환
x = data[0:,0:-1].astype('int64')
y = data[0:,-1]

# y분류 LabelEncoder로 숫자로 변경(0-건전, 1-부실)
encoder=LabelEncoder()
encoder.fit(y)
y_encode = encoder.transform(y)
```

```python
print("x:",x)
print("y:",y_encode)

print("x.shape:",x.shape) #(150,135)
print("y.shape:",y.shape) #(150,)
```

```python
#1. 데이터
x_train,x_test,y_train,y_test=train_test_split(x,y_encode,test_size=0.3,random_state=60)

print(x_train.shape)
print(x_test.shape)
print(y_train)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
#2. 모델
model = XGBClassifier(n_estimators=1000)
#3. 훈련, 평가예측
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
acc = model.score(x_test,y_test)
print("acc:",acc)
```

```python
feature_importances_ = model.feature_importances_
```

```python
data = pd.read_csv("data_company22.csv",index_col=0, header=0, thousands=',',encoding='cp949')
```

```python
data['건전성 여부'] = data['건전성 여부'].replace('부실',1).replace('건전',0)
```

```python
corr = data.corr()
```

```python
Survived = corr.loc[:,['건전성 여부']]
Survived = abs(Survived)
corelation = Survived.sort_values(by=['건전성 여부'],
                              ascending=False)
corelation = corelation[2:]
corelation
```

```python
corelation.to_csv('./카카오뱅크/corelation.csv')
```

```python
data_list = list(data)
data_list_0 = data_list[:39] # 49개임
```

```python
df = pd.DataFrame(columns=['변수','점수'])
for i in data_list_0 :
    try:
        cor = corelation.loc[i][0]+corelation.loc[i+'.1'][0]+corelation.loc[i+'.2'][0]
        importance = feature_importances.loc[i][0]+feature_importances.loc[i+'.1'][0]+feature_importances.loc[i+'.2'][0]
        c = cor + (importance*2)
        dict = {'변수':[i],'점수':[c]}
        print(c)
        df = df.append(pd.DataFrame(dict))
    except: continue
df
```

```python
score = df.sort_values(by=['점수'],
                              ascending=False)
```
